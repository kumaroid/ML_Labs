````python
int_types = [np.int8, np.int16, np.int32, np.int64]
float_types = [np.float16, np.float32, np.float64]

for int_type in int_types:
    info = np.iinfo(int_type)
    print(f"{int_type.__name__}: min = {info.min}, max = {info.max}")

for float_type in float_types:
    info = np.finfo(float_type)
    print(f"{float_type.__name__}: min = {info.min}, max = {info.max}")
````

2. Сбалансировано ли соотношение мужчин и женщин в наборе данных?

````python
gender_counts = df['Gender'].value_counts()

print("Количество мужчин:", gender_counts.get('Male', 0))
print("Количество женщин:", gender_counts.get('Female', 0))

if gender_counts.get('Male', 0) == gender_counts.get('Female', 0):
    print("Соотношение мужчин и женщин сбалансировано.")
else:
    print("Соотношение мужчин и женщин не сбалансировано.")
````

3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
high_stress = df[df['Stress Levels'] > df['Stress Levels'].median()]
low_stress = df[df['Stress Levels'] <= df['Stress Levels'].median()]

def blood_pressure(bp_series):
    systolic_values = bp_series.str.split('/').str[0].astype(float)
    return systolic_values

high_stress_bp = blood_pressure(high_stress['Blood Pressure (s/d)']).mean()
low_stress_bp = blood_pressure(low_stress['Blood Pressure (s/d)']).mean()


if (high_stress_bp > low_stress_bp):
    print("Правда)")
else:
    print("Неправда(")
````

4. Правда ли, что уровень холестерина у мужчин и женщин отличается?

````python
cholesterol_male = df[df['Gender'] == 'Male']['Cholesterol Level (mg/dL)'].mean()
cholesterol_female = df[df['Gender'] == 'Female']['Cholesterol Level (mg/dL)'].mean()

if (cholesterol_male != cholesterol_female):
    print("Правда)")
else:
    print("Неправда(")
````

5. Правда ли, что с возрастом женщины имеют свойство оценивать свои когнитивные способности хуже, чем в юности (порог юности - 30 лет, женщина в возрасте - 60 лет+)

````python
womens = df[df['Gender'] == 'Female']

young = womens[womens['Age (years)'] <= 30]
old = womens[womens['Age (years)'] >= 60]

young_cognitive = young['Cognitive Function'].mean()
old_cognitive = old['Cognitive Function'].mean()


if old_cognitive < young_cognitive:
    print("Правда)")
else:
    print("Неправда")
````

6. Влияет ли физическая активность на вес в возрасте более 50 лет?

````python
data = df[df['Age (years)'] > 50]

low_activity_weight = data[data['Physical Activity Level'] == 'Low']['Weight (kg)'].mean()
moderate_activity_weight = data[data['Physical Activity Level'] == 'Moderate']['Weight (kg)'].mean()
high_activity_weight = data[data['Physical Activity Level'] == 'High']['Weight (kg)'].mean()

if high_activity_weight < low_activity_weight and high_activity_weight < moderate_activity_weight:
    print("Влияет")
else:
    print("Не влияет")
````

7. Сравните минимальный вес мужчин с высоким и низким заработком

````python
min_weight_high_income_male = df[(df['Gender'] == 'Male') & (df['Income Level'] == 'High')]['Weight (kg)'].min()
min_weight_low_income_male = df[(df['Gender'] == 'Male') & (df['Income Level'] == 'Low')]['Weight (kg)'].min()

if (min_weight_high_income_male > min_weight_low_income_male):
    print("Минимальный вес мужчин с высоким заработком больше веса мужчин с низким заработком")
else:
    print("Минимальный вес мужчин с высоким заработком меньше веса мужчин с низким заработком")
````

8. Правда ли, что у людей с индексом массы тела больше 75го перцентиля встречаются хронические заболевания чаще, чем у людей с индесом массы тела в переделах от 30 до 60 перцентиля?

````python
bmi_75_percentile = df['BMI'].quantile(0.75)
bmi_30_percentile = df['BMI'].quantile(0.30)
bmi_60_percentile = df['BMI'].quantile(0.60)

high_bmi = df[df['BMI'] > bmi_75_percentile]
high_bmi_chronic_disease_count = high_bmi[high_bmi['Chronic Diseases'].notnull()].shape[0]
high_bmi_total_count = high_bmi.shape[0]

mid_bmi = df[(df['BMI'] >= bmi_30_percentile) & (df['BMI'] <= bmi_60_percentile)]
mid_bmi_chronic_disease_count = mid_bmi[mid_bmi['Chronic Diseases'].notnull()].shape[0]
mid_bmi_total_count = mid_bmi.shape[0]

high_bmi_percentage = high_bmi_chronic_disease_count / high_bmi_total_count * 100
mid_bmi_percentage = mid_bmi_chronic_disease_count / mid_bmi_total_count * 100


if high_bmi_percentage > mid_bmi_percentage:
    print("Правда)")
else:
    print("Неправда(")
````

9. Правда ли, что люди с высоким уровнем дохода имеют больший уровень физической активности, чем люди с низким доходом

````python
high_income_level = df[df['Income Level'] == 'High']
low_income_level = df[df['Income Level'] == 'Low']

high_income_activity = high_income_level[high_income_level['Physical Activity Level'] == 'Low'].shape[0] / high_income_level.shape[0]
low_income_activity = low_income_level[low_income_level['Physical Activity Level'] == 'Low'].shape[0] / low_income_level.shape[0]


if high_income_activity > low_income_activity:
    print("Правда)")
else:
    print("Неправда(")
````

6. Имеет ли связь зрение и способность слышать звуки на высоких частотах? Ответьте на вопрос - какая зависимость между ними и предположите почему зависимость именно такая?

````python
plt.figure(figsize=(13, 7))
sns.scatterplot(x=df['Vision Sharpness'], y=df['Hearing Ability (dB)'], hue=df['Age (years)'], palette="magma")

bins = np.linspace(df['Vision Sharpness'].min(), df['Vision Sharpness'].max(), 30)

df['Vision Bin'] = pd.cut(df['Vision Sharpness'], bins)

mean_values = df.groupby('Vision Bin', observed=False).agg({
    'Vision Sharpness': 'mean',
    'Hearing Ability (dB)': 'mean'
}).reset_index()

first_x = mean_values['Vision Sharpness'].iloc[0]
last_x = mean_values['Vision Sharpness'].iloc[-1]

first_y = mean_values['Hearing Ability (dB)'].iloc[0]
last_y = mean_values['Hearing Ability (dB)'].iloc[-1]

plt.plot([first_x, last_x], [first_y, last_y], color='blue', linestyle='-', label='Линейная регрессия')

plt.title('Связь между зрением и способностью слышать высокие частоты')
plt.xlabel('Зрение')
plt.ylabel('Слух на высоких частотах (dB)')
plt.legend()

plt.show()

# Линейная зависимость
````
